### Column names



````python
import pandas as pd
from pathlib import Path

# Setup paths
PROJECT_ROOT = Path.cwd().parent
DATA_DIR = PROJECT_ROOT / "data"
TRAIN_FILE = DATA_DIR / "training_data_lowercase.csv"
TEST_FILE = DATA_DIR / "testing_data_lowercase_nolabels.csv"

# Load data with correct delimiter (tab-separated)
train_df = pd.read_csv(TRAIN_FILE, sep='\t', header=None, names=['target', 'text'])
test_df = pd.read_csv(TEST_FILE, sep='\t', header=None, names=['target', 'text'])

# Now define your column references
TEXT_COL  = "text"
TARGET_COL = "target"

# Assertions
assert TEXT_COL in train_df.columns, f"TEXT_COL='{TEXT_COL}' not found."
assert TEXT_COL in test_df.columns, f"TEXT_COL='{TEXT_COL}' not found in test."
assert TARGET_COL in train_df.columns, f"TARGET_COL='{TARGET_COL}' not found."
assert TARGET_COL in test_df.columns, f"TARGET_COL='{TARGET_COL}' not found in test."

print(f"Train shape: {train_df.shape}")
print(f"Test shape: {test_df.shape}")
print(f"\nTrain columns: {train_df.columns.tolist()}")
print(f"Test columns: {test_df.columns.tolist()}")

# Label distribution
display(train_df[TARGET_COL].value_counts(dropna=False))
````

##  Train/Validation split
We split once for fast iteration and fair model comparison.


````python

X = train_df[TEXT_COL]
y = train_df[TARGET_COL]
X_test = test_df[TEXT_COL]

X_train, X_val, y_train, y_val = train_test_split(
    X, y,
    test_size=0.2,
    random_state=RANDOM_STATE,
    stratify=y
)

print("Train:", X_train.shape, "Val:", X_val.shape)
````

##  Lemmatization
Model 1.5 : TF‑IDF + LinearSVC


````python
# Normalize the data and tokenize
import nltk
from nltk.stem import WordNetLemmatizer


nltk.download("wordnet")
nltk.download("omw-1.4")

lemmatizer = WordNetLemmatizer()

def lemmatize_tokenizer(text: str):
    # minimal tokenization so lemmatization can happen
    tokens = re.findall(r"[A-Za-z]+", str(text))
    return [lemmatizer.lemmatize(t.lower()) for t in tokens]
````

##  Model
TF‑IDF + LinearSVC - Basic


````python
# Added Lemmatizer to model

Model1 = Pipeline(steps=[
    ("tfidf", TfidfVectorizer(tokenizer=lemmatize_tokenizer,token_pattern=None, ngram_range=(1,2), min_df=2, max_df=0.95,  sublinear_tf=True)),
    ("clf", LinearSVC(C=1.2))
]) # Ngrams with min document frequency and max doc for detecting the word patterns

t0 = time.perf_counter()
Model1.fit(X_train, y_train)
fit_s = time.perf_counter() - t0

y_pred = Model1.predict(X_val)

print("Model 1: TFIDF + LinearSVC")
print("Fit seconds:", round(fit_s, 3))
print("Accuracy:", round(accuracy_score(y_val, y_pred), 4))
print("F1 (weighted):", round(f1_score(y_val, y_pred, average="weighted"), 4))
````

**Comments:**

Goal: Finding how well the linearSVC model performs with no preprocessing.

Keeping the same parameters as the baseline model except, is changed to C=1.2 (best validation accuracy).

With preprocessing removed we have a performance increase of 0.01 across the board very small but we are already scoring high here from LinearSVC + TF-IDF, possible that some lemmatization has helped due to its normalization technique.



````python
print(classification_report(y_val, y_pred))

cm = confusion_matrix(y_val, y_pred, labels=np.unique(y))
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=np.unique(y))
disp.plot()
plt.show()
````

 **Comments:**

So still very dominate with the diagonal here (yellow): **3377** class 0 (fake) correctly predicted as 0, and **3180** class 1 (true) correctly predicted as 1.

Misclassifications are even lower: **138** samples of true 0 were predicted as 1, and **136** samples of true 1 were predicted as 0.



Overall performance is very strong and consistent: accuracy is 0.96, and both classes have F1 ≈ 0.96, meaning precision and recall are almost perfectly well balanced and provide the best score in (my) test so far.


##   LinearSVC + **LogisticRegression** Classifier



````python
from sklearn.ensemble import VotingClassifier
from sklearn.svm import LinearSVC
from sklearn.linear_model import LogisticRegression


ensemble_pipe = Pipeline([
    ("tfidf", TfidfVectorizer(ngram_range=(1,2), min_df=2, max_df=0.95, sublinear_tf=True)),
    ("ensemble", VotingClassifier(
        estimators=[
            ("svc", LinearSVC(C=1.0, max_iter=2000)),
           ("lr", LogisticRegression(max_iter=2000, n_jobs=-1, random_state=RANDOM_STATE))
        ],
        voting='hard'  # Breaks on soft voting
    ))
])

ensemble_pipe.fit(X_train, y_train)
ensemble_pred = ensemble_pipe.predict(X_val)

# Print the comparison
print("=== Ensemble (SVC + LR) ===")
print(classification_report(y_val, ensemble_pred))

labels = np.unique(y_val)
cm = confusion_matrix(y_val, ensemble_pred, labels=labels)

disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=labels)
disp.plot(values_format="d")
plt.title("Confusion Matrix — Ensemble (SVC + LR)")
plt.show()
````

##   LinearSVC + LR(NoPrePro): Final Evaluation

Applying LogisticRegression to this Model shows that recall for fake news has not really improved (Class: 0) from 0.94 > 0.9 but Class (1) for real names is slightly worse, 0.96 > 0.94. Showing a redistribution of class results. So better with recall but sloppier precision.

While LinearSVC is 0.96 Class 0 and 0.95 class 1 for parallel with precision and recall.

F1 stayed the same for both results.

It shows that this ensemble attempt didn't really improve the overall result and that linear SVC is best as it is with its very high and consistent scores of 0.95 overall.
